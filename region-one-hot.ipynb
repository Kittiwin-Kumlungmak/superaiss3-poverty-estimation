```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('data/Region_with_tumbol_last.csv')
df.Region = df.Region.astype('category')
df['Region_code'] = df.Region.cat.codes
df
```

```python
region2code = dict(enumerate(df.Region.cat.categories))
region2code
```

```python
from sklearn.preprocessing import OneHotEncoder

ohe = OneHotEncoder()
region_code = ohe.fit_transform(df.Region_code.values.reshape(-1,1))
region_code_df = pd.DataFrame(region_code.toarray())
region_code_df.columns = list(region2code.values())
region_code_df.columns = [
    'region_BKK',
    'region_Central',
    'region_West',
    'region_East',
    'region_North_East',
    'region_North',
    'region_South']
region_code_df
```

```python
out_df = pd.concat([df, region_code_df], axis=1)
out_df
```

```python
out_df.to_csv('data/tumbol_region_code.csv', index=False)
```

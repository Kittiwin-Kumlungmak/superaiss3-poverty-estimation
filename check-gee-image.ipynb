```python
import pandas as pd
import numpy as np
import os
```

```python
tumbol_id = []
images = []
main_dir = 'data/image'

for file in os.listdir(main_dir):
    tumbol_id.append(file[:-4])
    images.append(file)
    
images_df = pd.DataFrame({'tumbol_ID': tumbol_id, 'images': images})
images_df['tumbol_ID'] = images_df['tumbol_ID'].astype('int')
```

```python
train_df = pd.read_csv('data/ss3-tpmap-poverty-rate-prediction-accessibility/tpmap65_train_accessibility.csv')['tumbol_ID']
train_df = pd.DataFrame(train_df)
train_df
```

```python
test_df = pd.read_csv('data/ss3-tpmap-poverty-rate-prediction-accessibility/tpmap65_test.csv')['tumbol_ID']
test_df = pd.DataFrame(test_df)
test_df
```

```python
train_test_df = pd.concat([train_df, test_df], ignore_index=True)
train_test_df = pd.DataFrame(train_test_df)
train_test_df
```

```python
train_test_df.to_csv('all-tumbol-id.csv', index=False)
```

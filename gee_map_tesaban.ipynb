```python
import pandas as pd
import os
```

```python
tumbol_id_df = pd.read_csv('data/all-tumbol-id.csv')
tumbol_id_df
```

```python
tumbol_df = pd.read_csv('data/tumbon_week1.csv')[['TUMBOL_ID', 'MAP_TUMBOL_ID']]
tumbol_df.columns = ['tumbol_ID', 'MAP_TUMBOL_ID']
tumbol_df
```

```python
gee_df = pd.read_csv('data/GEE-merge_all.csv')
gee_df
```

```python
tesaban_df = tumbol_df[tumbol_df['tumbol_ID'] != tumbol_df['MAP_TUMBOL_ID']]
tesaban_df.index = range(len(tesaban_df))
tesaban_df = tesaban_df.dropna()
tesaban_df = tesaban_df.merge(gee_df, left_on= 'MAP_TUMBOL_ID', right_on='tumbol_ID')
tesaban_df = tesaban_df.drop(['MAP_TUMBOL_ID','tumbol_ID_y'], axis=1)
col_name = list(tesaban_df.columns)
col_name[0] = 'tumbol_ID'
tesaban_df.columns = col_name
tesaban_df.index = tesaban_df.tumbol_ID
tesaban_df = tesaban_df.drop('tumbol_ID', axis=1)
tesaban_df
```

```python
gee_map_df = tumbol_df.merge(gee_df, on='tumbol_ID', how='left')
gee_map_df = gee_map_df.drop('MAP_TUMBOL_ID',axis=1)
gee_map_df.index = gee_map_df.tumbol_ID
gee_map_df = gee_map_df.drop('tumbol_ID', axis=1)

for i in tesaban_df.index:
    gee_map_df.loc[i] = tesaban_df.loc[i]
    
gee_map_df
```

```python
gee_NDVI = pd.read_csv('data/GEE-v6_dirty.csv')[['tumbol_ID', 'NDWI','NDBI','NDVI']]
gee_NDVI
```

```python
out_df = gee_map_df.merge(gee_NDVI, on='tumbol_ID', how='left')
out_df
```

```python
out_df.to_csv('data/gee_processed.csv')
```
